```python
# 6. Write a Python program that logs an error message to a log file when a division by zero exception occurs.

import logging
import os

log_file_name = "division_errors.log"

if os.path.exists(log_file_name):
    os.remove(log_file_name)

logging.basicConfig(
    filename=log_file_name,
    level=logging.ERROR,
    format='%(asctime)s - %(levelname)s - %(message)s'
)

def safe_divide(numerator, denominator):
    try:
        result = numerator / denominator
        print(f"The result of {numerator} / {denominator} is: {result}")
        return result
    except ZeroDivisionError:
        error_message = f"Attempted division by zero: {numerator} / {denominator}"
        logging.error(error_message)
        print(f"Error: {error_message}. Check '{log_file_name}' for details.")
        return None
    except Exception as e:
        error_message = f"An unexpected error occurred during division: {e}"
        logging.error(error_message)
        print(f"An unexpected error occurred: {error_message}. Check '{log_file_name}' for details.")
        return None

print("\n--- Running Division Tests ---")
safe_divide(10, 2)
safe_divide(5, 0)
safe_divide(10, 3)
safe_divide(7, 0)
safe_divide(15, 'a')

print(f"\n--- Division Tests Complete. Check '{log_file_name}' for error logs. ---")

if os.path.exists(log_file_name):
    print(f"\nContent of '{log_file_name}':")
    with open(log_file_name, 'r') as f:
        print(f.read())
else:
    print(f"Log file '{log_file_name}' not found.")
```

```python
# 8. Write a program to handle a file opening error using exception handling?

def open_and_read_file(filename):
    try:
        with open(filename, 'r') as file:
            content = file.read()
            print(f"Successfully read content from '{filename}':\n{content}")
            return content
    except FileNotFoundError:
        print(f"Error: The file '{filename}' was not found. Please check the file path and ensure it exists.")
        return None
    except PermissionError:
        print(f"Error: Permission denied to access '{filename}'. Check file permissions.")
        return None
    except IOError as e:
        print(f"An I/O error occurred while opening '{filename}': {e}")
        return None
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
        return None

print("--- Testing with an existing file ---")
open_and_read_file("my_example_file.txt")

print("\n--- Testing with a non-existent file ---")
open_and_read_file("non_existent_file_for_error_test.txt")

print("\n--- Testing with a file that might cause permission issues (if applicable, e.g., trying to read a directory as a file) ---")
open_and_read_file("/dev/null")
```

```python
# 12. Write a program that demonstrates using multiple except blocks to handle different types of exceptions?

def perform_operation(value1, value2, operation):
    try:
        if operation == 'divide':
            result = value1 / value2
            print(f"Division Result: {result}")
        elif operation == 'multiply':
            result = value1 * value2
            print(f"Multiplication Result: {result}")
        elif operation == 'add_list':
            result = list(value1) + list(value2)
            print(f"List Addition Result: {result}")
        else:
            raise ValueError("Invalid operation specified.")

    except ZeroDivisionError:
        print("Error: Cannot divide by zero!")
    except TypeError as e:
        print(f"Error: Invalid type for operation. Details: {e}")
    except ValueError as e:
        print(f"Error: Invalid value. Details: {e}")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

print("--- Test Case 1: Successful Division ---")
perform_operation(10, 2, 'divide')

print("\n--- Test Case 2: Division by Zero ---")
perform_operation(5, 0, 'divide')

print("\n--- Test Case 3: Type Error (multiplying int with string) ---")
perform_operation(5, "hello", 'multiply')

print("\n--- Test Case 4: Type Error (adding incompatible types as list) ---")
perform_operation(1, 2, 'add_list')

print("\n--- Test Case 5: Value Error (invalid operation) ---")
perform_operation(10, 2, 'subtract')

print("\n--- Test Case 6: Successful List Addition ---")
perform_operation([1, 2], [3, 4], 'add_list')
```

```python
# 14. Write a program that uses the logging module to log both informational and error messages.

import logging
import os

log_file = "info_and_error.log"

if os.path.exists(log_file):
    os.remove(log_file)

logging.basicConfig(
    filename=log_file,
    level=logging.INFO,
    format='%(asctime)s - %(levelname)s - %(message)s'
)

def process_data(data):
    logging.info(f"Starting data processing for: {data}")
    try:
        if not isinstance(data, (int, float)) or data <= 0:
            raise ValueError("Invalid data for processing.")
        result = 100 / data
        logging.info(f"Successfully processed data '{data}'. Result: {result}")
        return result
    except ValueError as ve:
        logging.error(f"Data processing failed for '{data}': {ve}")
        return None
    except ZeroDivisionError:
        logging.error(f"Data processing failed for '{data}': Division by zero attempted.")
        return None
    except Exception as e:
        logging.error(f"An unexpected error occurred during processing data '{data}': {e}")
        return None

print("--- Running data processing tests ---")
process_data(5)
process_data(0)
process_data(20)
process_data("text")
process_data(-10)

print(f"\n--- Data processing tests complete. Check '{log_file}' for logs. ---")

if os.path.exists(log_file):
    print(f"\nContent of '{log_file}':")
    with open(log_file, 'r') as f:
        print(f.read())
else:
    print(f"Log file '{log_file}' not found.")
```

```python
# 19. Write a program that handles both IndexError and KeyError using a try-except block?

def access_data(data_list, data_dict, list_index, dict_key):
    try:
        print(f"Attempting to access list index {list_index}...")
        list_value = data_list[list_index]
        print(f"Value at index {list_index}: {list_value}")

        print(f"\nAttempting to access dictionary key '{dict_key}'...")
        dict_value = data_dict[dict_key]
        print(f"Value for key '{dict_key}': {dict_value}")

    except (IndexError, KeyError) as e:
        print(f"Error caught: {e}. The operation failed because of an invalid index or missing key.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

my_list = [10, 20, 30]
my_dict = {"name": "Alice", "age": 30}

print("--- Test Case 1: All accesses successful ---")
access_data(my_list, my_dict, 1, "name")

print("\n--- Test Case 2: IndexError (invalid list index) ---")
access_data(my_list, my_dict, 5, "age")

print("\n--- Test Case 3: KeyError (non-existent dictionary key) ---")
access_data(my_list, my_dict, 0, "country")

print("\n--- Test Case 4: Both errors possible (first one encountered will be caught) ---")
access_data(my_list, my_dict, 5, "country")
```

```python
# 21. Write a Python program that reads a file and prints the number of occurrences of a specific word?

def count_word_occurrences(filepath, search_word, case_sensitive=False):
    """
    Reads a file and counts the occurrences of a specific word.

    Args:
        filepath (str): The path to the file.
        search_word (str): The word to search for.
        case_sensitive (bool): If True, the search is case-sensitive. Default is False.

    Returns:
        int: The number of occurrences of the word, or -1 if the file cannot be read.
    """
    count = 0
    try:
        with open(filepath, 'r') as file:
            content = file.read()

            if not case_sensitive:
                content = content.lower()
                search_word = search_word.lower()

            words = content.split()
            for word in words:
                if word == search_word:
                    count += 1

        print(f"The word '{search_word}' appears {count} time(s) in '{filepath}'.")
        return count

    except FileNotFoundError:
        print(f"Error: The file '{filepath}' was not found.")
        return -1
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
        return -1

sample_file_path = "sample_text.txt"
with open(sample_file_path, 'w') as f:
    f.write("Hello world! This is a test file. Hello again. hello world. World.")

print("--- Test Case 1: Case-insensitive search for 'hello' ---")
count_word_occurrences(sample_file_path, "hello")

print("\n--- Test Case 2: Case-sensitive search for 'Hello' ---")
count_word_occurrences(sample_file_path, "Hello", case_sensitive=True)

print("\n--- Test Case 3: Case-insensitive search for 'world' ---")
count_word_occurrences(sample_file_path, "world")

print("\n--- Test Case 4: Search for a non-existent word ---")
count_word_occurrences(sample_file_path, "python")

print("\n--- Test Case 5: Search in a non-existent file ---")
count_word_occurrences("non_existent_file.txt", "word")
```

```python
# 22. How can you check if a file is empty before attempting to read its contents?

import os

def check_and_read_file(filepath):
    print(f"\n--- Checking file: '{filepath}' ---")
    if not os.path.exists(filepath):
        print(f"Error: The file '{filepath}' was not found.")
        return

    file_size = os.path.getsize(filepath)

    if file_size == 0:
        print(f"The file '{filepath}' is empty (size: 0 bytes).")
    else:
        print(f"The file '{filepath}' is not empty (size: {file_size} bytes). Reading content...")
        try:
            with open(filepath, 'r') as file:
                content = file.read()
                print("File content:")
                print(content)
        except IOError as e:
            print(f"Error reading file '{filepath}': {e}")
        except Exception as e:
            print(f"An unexpected error occurred: {e}")

file_with_content = "file_with_content_check.txt"
with open(file_with_content, 'w') as f:
    f.write("This file has some text.\nAnd another line for testing.")

empty_file_check = "empty_file_check.txt"
with open(empty_file_check, 'w') as f:
    pass

check_and_read_file(file_with_content)
check_and_read_file(empty_file_check)
check_and_read_file("non_existent_file_to_check.txt")
```

```python
# 23. Write a Python program that writes to a log file when an error occurs during file handling.

import logging
import os

file_handling_log_file = "file_handling_errors.log"

if os.path.exists(file_handling_log_file):
    os.remove(file_handling_log_file)

logging.basicConfig(
    filename=file_handling_log_file,
    level=logging.ERROR,
    format='%(asctime)s - %(levelname)s - %(message)s'
)

def safe_file_read(filepath):
    try:
        logging.info(f"Attempting to read file: {filepath}")
        with open(filepath, 'r') as f:
            content = f.read()
            print(f"Successfully read from '{filepath}'. Content snippet: '{content[:50]}...'\n")
            return content
    except FileNotFoundError:
        error_message = f"File not found: The file '{filepath}' does not exist."
        logging.error(error_message)
        print(f"Error: {error_message}. Check '{file_handling_log_file}' for details.")
        return None
    except PermissionError:
        error_message = f"Permission denied: Cannot access '{filepath}'. Check file permissions."
        logging.error(error_message)
        print(f"Error: {error_message}. Check '{file_handling_log_file}' for details.")
        return None
    except IOError as e:
        error_message = f"I/O error during reading '{filepath}': {e}"
        logging.error(error_message)
        print(f"Error: {error_message}. Check '{file_handling_log_file}' for details.")
        return None
    except Exception as e:
        error_message = f"An unexpected error occurred while handling '{filepath}': {e}"
        logging.error(error_message)
        print(f"Error: {error_message}. Check '{file_handling_log_file}' for details.")
        return None

existing_file = "my_example_file.txt"
if not os.path.exists(existing_file):
    with open(existing_file, 'w') as f:
        f.write("This is a test line for existing file.\nAnother line.")
safe_file_read(existing_file)

non_existent_file = "non_existent_file.txt"
safe_file_read(non_existent_file)

restricted_path = "/root"
if os.path.exists(restricted_path):
    print(f"\nAttempting to read a restricted path: {restricted_path}")
    safe_file_read(restricted_path)
else:
    print(f"\nSkipping test for restricted path '{restricted_path}' as it does not exist or is not suitable on this system.")

print(f"\n--- File handling tests complete. Check '{file_handling_log_file}' for error logs. ---")

if os.path.exists(file_handling_log_file):
    print(f"\nContent of '{file_handling_log_file}':")
    with open(file_handling_log_file, 'r') as f:
        print(f.read())
else:
    print(f"Log file '{file_handling_log_file}' not found.")
```
